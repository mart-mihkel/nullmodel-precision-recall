simulate and plot one nullmodel

````python
import numpy as np

from plotnine import *
from sklearn.metrics import precision_recall_curve
````

````python
y_true, y_scores = simulate_nullmodels(1, 100, acc=0.4, tresh=0.99)
curves = [precision_recall_curve(y_true=y_true, y_score=y_score)[:2] for y_score in y_scores]
p, r = np.hstack(curves)

ggplot() + aes(r, p) + geom_step()
````
